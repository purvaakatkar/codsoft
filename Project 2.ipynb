```python
#to balance the data we will perform undersampling on the data since the ml model will get cofused with the imbalanced dataset
# building a sample dataset containing similar distribution of both the transactions

nonfraud_sample = nonfraud.sample(n=492)
```

```python
newdata = pd.concat([nonfraud_sample,fraud],axis=0)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score,f1_score
```

```python
X = newdata.drop(columns='Class',axis=1)
Y = newdata['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X.shape,xtrain.shape,xtest.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
pred  = model.predict(xtest)
```

```python
print("Precision score: ",precision_score(ytest,pred))
print("Recall score: ",recall_score(ytest,pred))
print("F1 score: ",f1_score(ytest,pred))
```

```python
#evaluating using random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
import matplotlib.pyplot as plt
importances = rf.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 6))
plt.title('Feature Importances')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), [features[i] for i in indices], rotation=90)
plt.show()
```

```python
rf.fit(xtrain,ytrain)
```

```python
rf_pred = rf.predict(xtest)
```

```python
print("Precision score: ",precision_score(ytest,rf_pred))
print("Recall score: ",recall_score(ytest,rf_pred))
print("F1 score: ",f1_score(ytest,rf_pred))
```

```python
from sklearn.metrics import precision_recall_curve, auc
```

```python
y_pred_proba = rf.predict_proba(xtest)[:, 1]
```

```python
# Calculate precision-recall curve
precision, recall, _ = precision_recall_curve(ytest, y_pred_proba)
pr_auc = auc(recall, precision)

# Plotting the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', lw=2, label=f'PR curve (area = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()
```
